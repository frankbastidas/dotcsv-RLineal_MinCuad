```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
```

```python
#cargar la libreria
boston= load_boston()

print(boston.DESCR)
```

Formula erro cuadratico medio(MCO): $\beta = (X^TX)^{-1} X^TY$

```python
X=np.array(boston.data[:,5])
Y=np.array(boston.target)

plt.scatter(X,Y,alpha=0.3)

#se necesita agregar una columna de Unos para el termino independiente
X=np.array([np.ones(506),X]).T

#agregamos la formula para beta (@: multiplicacion matricial)
B=np.linalg.inv(X.T @ X) @ X.T @ Y

#beta B[0]: termino independiente(Y cuando X=0), B[1]: pendiente de la recta.
print(B)

#ecuacion de la recta y=mx+b
plt.plot([3, 10], [B[1] * 3 + B[0], B[1] * 10 + B[0]], c = "red")
plt.show()
```
